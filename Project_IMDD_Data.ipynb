### **Create a normalized form whose values range exactly between 0 and 1 so that the minimum has value 0 and maximum has value 1.**
Normalization usually means to scale a variable to have a values between 0 and 1, while standardization transforms data to have a mean of zero and a standard deviation of 1. This standardization is called a z-score, and data points can be standardized with the following formula: A z-score standardizes variables.

#### Data Normalization

```python
data_matrix = data.select_dtypes(['int64', 'float64'])
data_normalised = (data_matrix - data_matrix.min()) / (data_matrix.max() - data_matrix.min())
```

```python
data_normalised.head()
```

#### Data Standarization

```python
data_standarization = (data_matrix - data_matrix.mean()) / data_matrix.std()
```

```python
data_standarization.std()
```

```python
data_standarization.mean()
```

### **Compute the Softmax score**

```python
def softmax(vector):
    softmax = np.exp(vector)
    return softmax / softmax.sum()
```

```python
# convert list of numbers to a list of probabilities
softmax = softmax(data_normalised)
softmax
```

```python
np.sum(softmax)
```

#### Fix Ratings (Not Rated, Unrated, G, PG, R, PG-13, M, X, GP, NC-17)
The certificates section records the ratings certificates given to titles within each country.
- https://help.imdb.com/article/contribution/titles/certificates/GU757M8ZJ9ZPXB39?ref_=helpart_nav_27#

<b>Australia</b>
- <b>G</b> - General
- <b>PG</b> - Parental Guidance for under 15
- <b>M</b> - 15+ Recommended
- <b>R</b> - Passed only for persons 18 and over
- <b>X</b> - Sexually Explicit, 18s and over (Possibly only videos)

<b>Canada</b>
- <b>G</b> - Suitable for all
- <b>PG</b> - Parental Guidance
- <b>R</b> - Passed only for persons 18 and over
- <b>PG-13</b> - Teens
- <b>NC-17</b> - Adult

<b>USA</b>
- <b>Approved</b> - Pre-1968 titles only (from the MPA site) Under the Hays Code, films were simply approved or disapproved based on whether they were deemed 'moral' or 'immoral'.)

etc...

Here we can find the ratings we need to make the data more clear for the audience. You can find more information at the link above.

```python
data['Rating'].replace('G','General', inplace=True)
data['Rating'].replace('PG','Parental Guidance', inplace=True)
data['Rating'].replace('GP','Parental Guidance', inplace=True)
data['Rating'].replace('R','Over 18', inplace=True)
data['Rating'].replace('PG-13','Teens', inplace=True)
data['Rating'].replace('M','Teens', inplace=True)
data['Rating'].replace('NC-17','Adult', inplace=True)
data['Rating'].replace('X','Sexually Explicit', inplace=True)
data['Rating'].replace('Not Rated','Unrated', inplace=True)
```

# Type of data analysis
## Univariate
> is a term commonly used in statistics to describe a type of data which consists of observations on only a single characteristic or attribute.

>Univariate analysis is the simplest form of analyzing data. “Uni” means “one”, so in other words your data has only one variable.

## Bivariate
>In statistics, bivariate data is data on each of two variables, where each value of one of the variables is paired with a value of the other variable. Typically it would be of interest to investigate the possible association between the two variables.

>Bivariate analysis is one of the simplest forms of quantitative (statistical) analysis. It involves the analysis of two variables (often denoted as X, Y), for the purpose of determining the empirical relationship between them.

## Multivariate
>Multivariate statistics is a subdivision of statistics encompassing the simultaneous observation and analysis of more than one outcome variable. The application of multivariate statistics is multivariate analysis.

>Multivariate analysis is based on the principles of multivariate statistics, which involves observation and analysis of more than one statistical outcome variable at a time.
